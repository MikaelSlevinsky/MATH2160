```julia
using Images, Interact, Colors
```

```julia
p = load("/Users/Mikael/Documents/Courses/MATH2160/communism.jpg")
```

```julia
m,n = size(p.data);
R = Float64[p.data[i,j].r for i in 1:m, j in 1:n]
G = Float64[p.data[i,j].g for i in 1:m, j in 1:n]
B = Float64[p.data[i,j].b for i in 1:m, j in 1:n];
m,n
```

```julia
R
```

```julia
UR,ΣR,VR = svd(R);
UG,ΣG,VG = svd(G);
UB,ΣB,VB = svd(B);
```

```julia
@manipulate for r in 0:100
    RR = UR[:,1:r]*(ΣR[1:r].*VR[:,1:r]')
    GG = UG[:,1:r]*(ΣG[1:r].*VG[:,1:r]')
    BB = UB[:,1:r]*(ΣB[1:r].*VB[:,1:r]')
    p.data[:] = [RGB4{U8}(max(min(RR[i,j],1.0),0.0),max(min(GG[i,j],1.0),0.0),max(min(BB[i,j],1.0),0.0)) for i in 1:m, j in 1:n]
end
```

```julia
@manipulate for rR in 0:100, rG in 0:100, rB in 0:100
    RR = UR[:,1:rR]*(ΣR[1:rR].*VR[:,1:rR]')
    GG = UG[:,1:rG]*(ΣG[1:rG].*VG[:,1:rG]')
    BB = UB[:,1:rB]*(ΣB[1:rB].*VB[:,1:rB]')
    p.data[:] = [RGB4{U8}(max(min(RR[i,j],1.0),0.0),max(min(GG[i,j],1.0),0.0),max(min(BB[i,j],1.0),0.0)) for i in 1:m, j in 1:n]
end
```
